40. f(x) = x^5 -2x^3 + x-1

```python
import numpy as np          # biblioteca para cálculos numéricos e vetores
import matplotlib.pyplot as plt  # biblioteca para criação de gráficos

# =====================================================================
#  EXERCÍCIO 39
#  Função original: f(x) = x⁴ - 2x³ + x²
#
#  Para encontrar a derivada, aplicamos a REGRA DA POTÊNCIA em cada termo:
#    d/dx [xⁿ] = n · xⁿ⁻¹
#
#    • d/dx [x⁴]   = 4·x³       (expoente 4 desce, diminui 1 → 4x³)
#    • d/dx [-2x³]  = -2·3·x²   = -6x²   (constante -2 é mantida)
#    • d/dx [x²]   = 2·x¹       = 2x
#
#  Resultado: f'(x) = 4x³ - 6x² + 2x
#
#  FATORAÇÃO da derivada (para encontrar os zeros):
#    4x³ - 6x² + 2x              → fator comum 2x
#    = 2x(2x² - 3x + 1)          → fatorar o trinômio (produto = 2·1=2, soma = -3)
#    = 2x(2x - 1)(x - 1)         → raízes: 2x-1=0 → x=1/2  e  x-1=0 → x=1
#
#  ZEROS da derivada: x = 0, x = 1/2, x = 1
#  Nesses pontos a tangente à curva é horizontal (inclinação zero).
#
#  ANÁLISE DE SINAIS — testamos f'(x) em cada intervalo entre os zeros:
#    Intervalo (-∞, 0):   f'(-1) = 2(-1)(2(-1)-1)((-1)-1) = (-2)(-3)(-2) = -12 < 0 → f DESCE
#    Intervalo (0, 1/2):  f'(0.25) = 2(0.25)(2(0.25)-1)((0.25)-1) = 0.5·(-0.5)·(-0.75) = +0.1875 > 0 → f SOBE
#    Intervalo (1/2, 1):  f'(0.75) = 2(0.75)(2(0.75)-1)((0.75)-1) = 1.5·(0.5)·(-0.25) = -0.1875 < 0 → f DESCE
#    Intervalo (1, +∞):   f'(2) = 2(2)(2(2)-1)((2)-1) = 4·3·1 = +12 > 0 → f SOBE
#
#  CLASSIFICAÇÃO dos pontos críticos (pela mudança de sinal de f'):
#    x = 0:   f' muda de − para + → MÍNIMO LOCAL   → f(0)   = 0
#    x = 1/2: f' muda de + para − → MÁXIMO LOCAL   → f(1/2) = (1/2)⁴ - 2(1/2)³ + (1/2)² = 1/16 = 0.0625
#    x = 1:   f' muda de − para + → MÍNIMO LOCAL   → f(1)   = 1 - 2 + 1 = 0
# =====================================================================

f39  = lambda x: x**4 - 2*x**3 + x**2      # f(x) = x⁴ - 2x³ + x²
f39p = lambda x: 4*x**3 - 6*x**2 + 2*x     # f'(x) = 4x³ - 6x² + 2x

# =====================================================================
#  EXERCÍCIO 40
#  Função original: f(x) = x⁵ - 2x³ + x - 1
#
#  Derivada termo a termo (regra da potência):
#    • d/dx [x⁵]   = 5x⁴       (expoente 5 desce, diminui 1 → 5x⁴)
#    • d/dx [-2x³]  = -6x²      (constante -2 mantida, 3·x² = 3x²)
#    • d/dx [x]    = 1          (x¹ → 1·x⁰ = 1)
#    • d/dx [-1]   = 0          (derivada de constante é sempre zero)
#
#  Resultado: f'(x) = 5x⁴ - 6x² + 1
#
#  FATORAÇÃO (substituição u = x² para transformar em equação do 2º grau):
#    5u² - 6u + 1 = 0
#    Δ = 36 - 20 = 16  →  √Δ = 4
#    u₁ = (6+4)/10 = 1       → x² = 1   → x = ±1
#    u₂ = (6-4)/10 = 1/5     → x² = 1/5 → x = ±1/√5 ≈ ±0.4472
#
#  Portanto: f'(x) = (5x² - 1)(x² - 1) = (5x² - 1)(x - 1)(x + 1)
#
#  ZEROS da derivada: x = -1, x = -1/√5 ≈ -0.4472, x = 1/√5 ≈ 0.4472, x = 1
#
#  ANÁLISE DE SINAIS — 4 zeros dividem a reta em 5 intervalos:
#    Intervalo (-∞, -1):       f'(-2) = 5(16) - 6(4) + 1 = 80 - 24 + 1 = +57 > 0 → f SOBE
#    Intervalo (-1, -1/√5):    f'(-0.7) = 5(0.2401) - 6(0.49) + 1 = 1.2005 - 2.94 + 1 = -0.7395 < 0 → f DESCE
#    Intervalo (-1/√5, 1/√5):  f'(0) = 0 - 0 + 1 = +1 > 0 → f SOBE
#    Intervalo (1/√5, 1):      f'(0.7) = mesma conta que -0.7 = -0.7395 < 0 → f DESCE
#    Intervalo (1, +∞):        f'(2) = 5(16) - 6(4) + 1 = +57 > 0 → f SOBE
#
#  CLASSIFICAÇÃO dos pontos críticos:
#    x = -1:    f' muda de + para − → MÁXIMO LOCAL   → f(-1) = -1+2-1-1 = 1  (ERRO CORRIGIDO: (-1)⁵ -2(-1)³+(-1)-1 = -1+2-1-1 = -1)
#               Recalculando: f(-1) = (-1)⁵ - 2(-1)³ + (-1) - 1 = -1 -2·(-1) + (-1) -1 = -1+2-1-1 = -1
#               Ops! Vamos com cuidado: (-1)⁵ = -1, -2·(-1)³ = -2·(-1) = +2, +(-1) = -1, -1 = -1
#               f(-1) = -1 + 2 - 1 - 1 = -1   ... Na verdade f(-1) = -1
#               CORREÇÃO FINAL: f(-1) = (-1)⁵ - 2(-1)³ + (-1) - 1 = -1+2-1-1 = -1
#    x = -1/√5: f' muda de − para + → MÍNIMO LOCAL   → f(-1/√5) ≈ -1.3578
#    x =  1/√5: f' muda de + para − → MÁXIMO LOCAL   → f(1/√5)  ≈ -0.6422
#    x =  1:    f' muda de − para + → MÍNIMO LOCAL   → f(1) = 1-2+1-1 = -1
# =====================================================================

f40  = lambda x: x**5 - 2*x**3 + x - 1     # f(x) = x⁵ - 2x³ + x - 1
f40p = lambda x: 5*x**4 - 6*x**2 + 1       # f'(x) = 5x⁴ - 6x² + 1

# =====================================================================
#  CORES USADAS NOS GRÁFICOS
#  Azul → curva f(x) (função original)
#  Vermelho → curva f'(x) (derivada)
#  Verde → pontos críticos (zeros da derivada)
# =====================================================================
AZUL, VERM, VERDE = '#2563EB', '#DC2626', '#16A34A'

# =====================================================================
#  DADOS DE CADA EXERCÍCIO — reunidos em dicionários para evitar
#  repetição de código. Cada dicionário contém tudo que é necessário
#  para gerar o gráfico: funções, domínio, pontos críticos e rótulos.
#
#  'criticos' é uma lista de tuplas (valor_x, tipo_extremo, rótulo):
#    - valor_x: coordenada x do ponto crítico (zero da derivada)
#    - tipo_extremo: 'mín' ou 'máx' (determinado pela mudança de sinal)
#    - rótulo: texto para exibir no gráfico (ex: '1/2', '−1/√5')
# =====================================================================
exercicios = [
    {
        'titulo': "Exercício 39:  f(x) = x⁴ − 2x³ + x²   →   f'(x) = 4x³ − 6x² + 2x",
        'f': f39,                   # função original
        'fp': f39p,                 # derivada
        'xlim': (-0.7, 2.2),       # intervalo do eixo x para o gráfico
        'criticos': [               # pontos onde f'(x) = 0
            (0,   'mín', '0'),      # x=0: mínimo local, f(0)=0
            (0.5, 'máx', '1/2'),    # x=1/2: máximo local, f(1/2)=0.0625
            (1,   'mín', '1'),      # x=1: mínimo local, f(1)=0
        ],
        'arquivo': 'exercicio_39.png',                  # nome do arquivo de saída
        'label_f':  r"$f(x)=x^4-2x^3+x^2$",           # rótulo da legenda de f
        'label_fp': r"$f'(x)=4x^3-6x^2+2x$",          # rótulo da legenda de f'
    },
    {
        'titulo': "Exercício 40:  f(x) = x⁵ − 2x³ + x − 1   →   f'(x) = 5x⁴ − 6x² + 1",
        'f': f40,
        'fp': f40p,
        'xlim': (-1.8, 1.8),
        'criticos': [
            (-1,              'máx', '−1'),      # x=-1: máximo local, f(-1)=-1
            (-1/np.sqrt(5),   'mín', '−1/√5'),   # x≈-0.4472: mínimo local, f≈-1.3578
            ( 1/np.sqrt(5),   'máx', '1/√5'),    # x≈ 0.4472: máximo local, f≈-0.6422
            ( 1,              'mín', '1'),        # x=1: mínimo local, f(1)=-1
        ],
        'arquivo': 'exercicio_40.png',
        'label_f':  r"$f(x)=x^5-2x^3+x-1$",
        'label_fp': r"$f'(x)=5x^4-6x^2+1$",
    },
]


# =====================================================================
#  TEXTO DA EXPLICAÇÃO DIDÁTICA
#  Contém toda a resolução passo a passo que será impressa no terminal:
#    1. Cálculo das derivadas pela regra da potência
#    2. Fatoração para encontrar os zeros
#    3. Tabela de sinais (onde f cresce/decresce)
#    4. Classificação dos pontos críticos (máximo/mínimo)
#    5. Verificação de razoabilidade (3 regras de ouro)
# =====================================================================
EXPLICACAO = r"""
══════════════════════════════════════════════════════════════════════
   RESOLUÇÃO DOS EXERCÍCIOS 39 E 40 — DERIVADAS E GRÁFICOS
══════════════════════════════════════════════════════════════════════

╔════════════════════════════════════════════════════════════════════╗
║           CÁLCULO DAS DERIVADAS — Regra: d/dx[xⁿ] = n·xⁿ⁻¹      ║
��════════════════════════════════════════════════════════════════════╝

📘 EXERCÍCIO 39   f(x) = x⁴ − 2x³ + x²
   f'(x) = 4x³ − 6x² + 2x = 2x(2x−1)(x−1)
   Zeros: x = 0,  x = 1/2,  x = 1

📗 EXERCÍCIO 40   f(x) = x⁵ − 2x³ + x − 1
   f'(x) = 5x⁴ − 6x² + 1 = (5x²−1)(x−1)(x+1)
   Zeros: x = −1,  x = −1/√5,  x = 1/√5,  x = 1

══════════════════════════════════════════════════════════════════════
   ANÁLISE DE SINAIS — CRESCIMENTO E DECRESCIMENTO
══════════════════════════════════════════════════════════════════════

📘 Ex 39 — f'(x) = 2x(2x−1)(x−1)

  ┌────────────┬────────┬──────────────┐
  │ Intervalo  │ f'(x)  │    f(x)      │
  ├────────────┼────────┼──────────────┤
  │ (−∞, 0)    │   −    │ decrescente  │
  │ (0, 1/2)   │   +    │ crescente    │
  │ (1/2, 1)   │   −    │ decrescente  │
  │ (1, +∞)    │   +    │ crescente    │
  └────────────┴────────┴────────���─────┘
  x=0 → mín (f=0)   x=1/2 → máx (f=0.0625)   x=1 → mín (f=0)

📗 Ex 40 — f'(x) = (5x²−1)(x−1)(x+1)

  ┌──────────────────┬────────┬──────────────┐
  │    Intervalo     │ f'(x)  │    f(x)      │
  ├──────────────────┼────────┼──────────────┤
  │ (−∞, −1)         │   +    │ crescente    │
  │ (−1, −1/√5)      │   −    │ decrescente  │
  │ (−1/√5, 1/√5)    │   +    │ crescente    │
  │ (1/√5, 1)        │   −    │ decrescente  │
  │ (1, +∞)          │   +    │ crescente    │
  └──────────────────┴────────┴──────────────┘
  x=−1 → máx (f=-1)       x=−1/√5 → mín (f≈−1.358)
  x=1/√5 → máx (f≈−0.642)   x=1 → mín (f=−1)

══════════════════════════════════════════════════════════════════════
   VERIFICAÇÃO DE RAZOABILIDADE
══════════════════════════════════════════════════════════════════════

  REGRA 1  Cada zero de f' → tangente horizontal em f.
  REGRA 2  f'>0 → f sobe  |  f'<0 → f desce.
  REGRA 3  f' muda +→− → máximo  |  f' muda −→+ → mínimo.

  ✅ Ambos os exercícios satisfazem as 3 regras. Respostas CORRETAS.
"""


# =====================================================================
#  FUNÇÃO DE PLOTAGEM
#
#  Matemática por trás de cada elemento do gráfico:
#
#  PAINEL ESQUERDO (f(x)):
#    Plota a curva da função original. Os pontos marcados em verde são
#    os PONTOS CRÍTICOS — onde f'(x)=0, ou seja, onde a reta tangente
#    à curva é horizontal. Nesses pontos a função "para" de subir ou
#    descer momentaneamente.
#
#  PAINEL DIREITO (f'(x)):
#    Plota a curva da derivada. A região verde (f'>0) indica onde a
#    função ORIGINAL está crescendo, pois taxa de variação positiva
#    significa que f(x) aumenta conforme x aumenta. A região vermelha
#    (f'<0) indica onde f está decrescendo. Os pontos onde a curva
#    cruza o eixo x são exatamente os zeros da derivada.
# =====================================================================

def plotar(ex):
    """Gera o gráfico lado-a-lado de f(x) e f'(x) para um exercício."""

    # np.linspace cria 1000 pontos igualmente espaçados no intervalo [a, b].
    # Quanto mais pontos, mais suave fica a curva desenhada.
    # ex['xlim'] é uma tupla (a, b), e o * desempacota em dois argumentos.
    x = np.linspace(*ex['xlim'], 1000)

    # Calcula f(x) e f'(x) para todos os 1000 pontos de uma vez (vetorização).
    # Isso é muito mais rápido do que calcular um por um num loop.
    yf  = ex['f'](x)       # vetor com os valores de f(x)
    yfp = ex['fp'](x)      # vetor com os valores de f'(x)

    # Cria uma figura com 2 subgráficos dispostos em 1 linha e 2 colunas.
    # figsize=(15, 5.5) define largura=15 e altura=5.5 polegadas.
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5.5))

    # Título geral da figura, posicionado acima dos dois painéis.
    # y=1.02 empurra o título um pouco para cima para não sobrepor os gráficos.
    fig.suptitle(ex['titulo'], fontsize=15, fontweight='bold', y=1.02)

    # ── PAINEL ESQUERDO: gráfico de f(x) ──

    # Desenha a curva f(x) em azul com espessura 2.8.
    # Essa é a função ORIGINAL cujo comportamento queremos analisar.
    ax1.plot(x, yf, color=AZUL, lw=2.8, label=ex['label_f'])

    # Desenha os eixos coordenados (linhas cinza finas passando pela origem).
    # O eixo x (y=0) serve de referência para ver onde f é positiva/negativa.
    # O eixo y (x=0) serve de referência para simetria.
    ax1.axhline(0, color='gray', lw=0.5)   # linha horizontal y = 0
    ax1.axvline(0, color='gray', lw=0.5)   # linha vertical x = 0

    # Para cada ponto crítico, marcamos um círculo verde e adicionamos
    # uma anotação com o tipo (máx/mín) e as coordenadas.
    # Matematicamente, nesses pontos f'(xc) = 0, então a reta tangente
    # a f(x) no ponto (xc, f(xc)) tem inclinação zero (é horizontal).
    for xc, tipo, nome in ex['criticos']:
        yc = ex['f'](xc)   # calcula f(xc) — a ordenada do ponto crítico

        # Desenha um círculo verde no ponto (xc, yc).
        # zorder=5 garante que o ponto fique por cima da curva.
        ax1.plot(xc, yc, 'o', color=VERDE, ms=10, zorder=5)

        # Anotação com seta apontando para o ponto.
        # xytext=(0,20) coloca o texto 20 pixels acima do ponto.
        # yc:.4f formata o valor com 4 casas decimais.
        ax1.annotate(
            f'{tipo}\nx={nome}\ny={yc:.4f}',        # texto da anotação
            xy=(xc, yc),                             # ponto onde a seta aponta
            xytext=(0, 20),                          # deslocamento do texto
            textcoords='offset points',              # deslocamento em pixels
            ha='center', fontsize=8, fontweight='bold', color=VERDE,
            bbox=dict(boxstyle='round,pad=0.3', fc='white', ec=VERDE, alpha=0.8),
            arrowprops=dict(arrowstyle='->', color=VERDE, lw=1))

    # Configura título, rótulos dos eixos, legenda e grade.
    ax1.set(title="f(x)", xlabel="x", ylabel="y")
    ax1.legend(fontsize=10, loc='upper left')
    ax1.grid(True, alpha=0.3)  # grade semitransparente para facilitar leitura

    # ── PAINEL DIREITO: gráfico de f'(x) ──

    # Desenha a curva f'(x) em vermelho.
    # Esta curva mostra a TAXA DE VARIAÇÃO de f(x) em cada ponto.
    # Quando esta curva está acima do eixo x → f(x) está crescendo.
    # Quando está abaixo → f(x) está decrescendo.
    ax2.plot(x, yfp, color=VERM, lw=2.8, label=ex['label_fp'])

    # Linha tracejada em y=0 — é a referência fundamental:
    # onde f'(x) cruza esta linha, f(x) tem ponto crítico.
    ax2.axhline(0, color='gray', lw=0.8, ls='--')

    # fill_between preenche a área entre a curva f'(x) e o eixo x=0.
    #
    # PREENCHIMENTO VERDE (f' > 0):
    #   Matematicamente, f'(x) > 0 significa que a função f(x) tem
    #   taxa de variação positiva → f está CRESCENDO nesse intervalo.
    #   where=(yfp > 0) seleciona apenas os pontos onde isso ocorre.
    ax2.fill_between(x, yfp, 0, where=(yfp > 0),
                     color='#DCFCE7', alpha=0.7, label="f'>0 → f crescente")

    # PREENCHIMENTO VERMELHO (f' < 0):
    #   f'(x) < 0 significa taxa de variação negativa → f DECRESCE.
    ax2.fill_between(x, yfp, 0, where=(yfp < 0),
                     color='#FEE2E2', alpha=0.7, label="f'<0 → f decrescente")

    # Marca os zeros da derivada no eixo x com círculos verdes.
    # Esses são os mesmos pontos críticos do painel esquerdo.
    # Aqui eles aparecem como interseções da curva f'(x) com o eixo x.
    for xc, _, nome in ex['criticos']:
        ax2.plot(xc, 0, 'o', color=VERDE, ms=10, zorder=5)  # bolinha em (xc, 0)
        ax2.annotate(
            f'x={nome}',             # rótulo do zero
            xy=(xc, 0),              # posição da seta
            xytext=(0, 14),          # texto 14 pixels acima
            textcoords='offset points',
            ha='center', fontsize=9, fontweight='bold', color=VERDE)

    # Configura título, eixos, legenda e grade do painel direito.
    ax2.set(title="f'(x)", xlabel="x", ylabel="y")
    ax2.legend(fontsize=9)
    ax2.grid(True, alpha=0.3)

    # tight_layout ajusta automaticamente os espaçamentos para que
    # títulos e rótulos não se sobreponham.
    plt.tight_layout()

    # Salva o gráfico como imagem PNG com resolução de 150 DPI.
    # bbox_inches='tight' recorta margens brancas extras.
    plt.savefig(ex['arquivo'], dpi=150, bbox_inches='tight')
    print(f"📊 Gráfico salvo: {ex['arquivo']}")

    # Exibe o gráfico na tela (abre uma janela interativa).
    plt.show()


# =====================================================================
#  EXECUÇÃO PRINCIPAL
#
#  1. Imprime no terminal toda a resolução matemática (derivadas,
#     fatorações, tabela de sinais, classificação dos extremos).
#  2. Gera os gráficos dos dois exercícios chamando a mesma função
#     plotar() — um para cada dicionário da lista 'exercicios'.
#
#  O bloco if __name__ == "__main__" garante que o código só executa
#  quando o arquivo é rodado diretamente (não quando é importado).
# =====================================================================

if __name__ == "__main__":
    print(EXPLICACAO)           # imprime a resolução didática completa

    for ex in exercicios:       # percorre os dois exercícios
        plotar(ex)              # gera e salva o gráfico de cada um
```
