43. Um avião voando a uma velocidade constante de 300 km/h passa sobre uma estação de radar no solo a uma altitude de 1 km e su-bindo em um ângulo de 30°. A que taxa está crescendo a distân-cia do avião em relação à estação de radar 1 minuto mais tarde?

```python
import math

# Dados do problema
v = 300  # km/h (velocidade constante do avião)
h0 = 1   # km (altitude inicial)
alpha_deg = 30 # graus (ângulo de subida)
t_min = 1 # minutos (tempo após passar pela estação de radar)

# Converter unidades e ângulos
alpha_rad = math.radians(alpha_deg)
t_hr = t_min / 60 # converter minutos para horas

# Componentes da velocidade
v_h = v * math.cos(alpha_rad) # Componente horizontal da velocidade
v_v = v * math.sin(alpha_rad) # Componente vertical da velocidade

# Posição do avião no tempo t
x_at_t = v_h * t_hr # Distância horizontal percorrida
h_at_t = h0 + v_v * t_hr # Altitude no tempo t

# Distância do avião até a estação de radar no tempo t (D(t))
D_at_t = math.sqrt(x_at_t**2 + h_at_t**2)

# Taxa de variação da distância (dD/dt)
# Usando a fórmula: dD/dt = (x * dx/dt + h * dh/dt) / D
# Onde dx/dt = v_h e dh/dt = v_v
# E x = v_h * t, h = h0 + v_v * t

# Simplificando a expressão para dD/dt:
# dD/dt = ( (v_h^2 + v_v^2) * t + h0 * v_v ) / D
# Como v_h^2 + v_v^2 = v^2, temos:
rate_of_change = (v**2 * t_hr + h0 * v_v) / D_at_t

print(f"Velocidade do avião: {v} km/h")
print(f"Altitude inicial: {h0} km")
print(f"Ângulo de subida: {alpha_deg}°")
print(f"Tempo decorrido: {t_min} minuto(s)")
print(f"--------------------------------------")
print(f"Distância horizontal no tempo t: {x_at_t:.2f} km")
print(f"Altitude no tempo t: {h_at_t:.2f} km")
print(f"Distância do avião à estação de radar no tempo t: {D_at_t:.2f} km")
print(f"A taxa na qual a distância do avião em relação à estação de radar está crescendo é: {rate_of_change:.2f} km/h")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math # Adicionado para corrigir o NameError

# Definir um intervalo de tempo para o gráfico (por exemplo, de 0 a 5 minutos)
t_values_min = np.linspace(0, 5, 100) # 100 pontos entre 0 e 5 minutos
t_values_hr = t_values_min / 60 # Converter para horas

# Listas para armazenar as distâncias calculadas
distances = []

# Recalcular os parâmetros necessários (velocidade, ângulo)
# Estes já foram definidos anteriormente, mas é bom ter certeza
v = 300  # km/h
h0 = 1   # km
alpha_deg = 30 # graus
alpha_rad = math.radians(alpha_deg)

v_h = v * math.cos(alpha_rad)
v_v = v * math.sin(alpha_rad)

# Calcular D(t) para cada ponto no tempo
for t_hr_val in t_values_hr:
    x_val = v_h * t_hr_val
    h_val = h0 + v_v * t_hr_val
    D_val = math.sqrt(x_val**2 + h_val**2)
    distances.append(D_val)

# Criar o gráfico
fig = plt.figure(figsize=(10, 6))
plt.plot(t_values_min, distances, label='Distância D(t)')
plt.title('Variação da Distância do Avião à Estação de Radar ao Longo do Tempo')
plt.xlabel('Tempo (minutos)')
plt.ylabel('Distância (km)')
plt.grid(True)
plt.legend()
plt.show()
```
