## 1. Objetivo do laboratorio

Implementar um pequeno Sistema de Gestao do Conhecimento em Python cobrindo o ciclo:

- **Producao**: capturar e estruturar o conhecimento do especialista.
- **Armazenamento**: registrar o conhecimento em uma base organizada.
- **Disseminacao**: permitir a recuperacao por busca.
- **Aplicacao**: automatizar uma decisao simples com motor de regras.

O problema central do laboratorio e evitar a **amnesia corporativa**, preservando o conhecimento do Sr. Carlos antes de sua saida da empresa.


## 2. Contexto do caso

No cenario proposto, o Sr. Carlos e o unico tecnico que sabe resolver a lentidao do ERP nas segundas-feiras. A partir do relato verbal dele, o conhecimento tacito precisa ser convertido em um formato explicito para que a empresa consiga:

- preservar a experiencia adquirida;
- consultar a solucao de forma padronizada;
- usar esse conhecimento em futuras decisoes operacionais.


````python
from pathlib import Path
from pprint import pprint
from IPython.display import Markdown, display

from sgc_empresa import (
    buscar_conhecimento,
    converter_tacito_para_explicito,
    diagnostico_servidor,
    montar_base_exemplo,
    salvar_base_json,
)

LAB_DIR = Path('.').resolve()

def mostrar_codigo(nome_arquivo):
    caminho = LAB_DIR / nome_arquivo
    codigo = caminho.read_text(encoding='utf-8')
    display(Markdown(f"### Codigo-fonte: `{nome_arquivo}`"))
    display(Markdown(f"```python\n{codigo}\n```"))
````

## 3. Conversao do conhecimento tacito em explicito

A primeira tarefa do SGC e estruturar o relato verbal do especialista em um **Procedimento Operacional Padrao (POP)**. Esse e o ponto exato em que o conhecimento tacito e transformado em conhecimento explicito.


````python
pop = converter_tacito_para_explicito()
pprint(pop, sort_dicts=False)
````

### Analise

O POP acima registra de forma objetiva:

- o problema observado;
- a causa raiz identificada;
- o passo a passo da resolucao;
- o indicador de erro fisico critico.

Com isso, a empresa deixa de depender exclusivamente da memoria do especialista.


## 4. Parte 1 - Producao e armazenamento

Nesta etapa, a base inicial de conhecimento e criada e depois enriquecida com novos registros. A persistencia em JSON garante que o conteudo fique salvo fora da memoria do programa.


````python
base = montar_base_exemplo()
pprint(base, sort_dicts=False)

arquivo = salvar_base_json(base)
print(f"\nBase salva em: {arquivo}")
````

### Evidencia de execucao da Parte 1

Ao exportar este notebook para PDF, a saida da celula anterior funciona como evidencia organizada da execucao, mostrando tanto a estrutura da base quanto a gravacao do arquivo `base_conhecimento.json`.


## 5. Parte 2 - Disseminacao do conhecimento

O valor do conhecimento armazenado depende da facilidade de recuperacao. Para isso, a funcao de busca usa **list comprehension**, que filtra a base por palavra-chave no titulo, na causa ou na solucao.

Forma geral da estrutura usada:

`[item for item in base if condicao]`

No laboratorio, essa estrategia torna a busca mais curta, legivel e adequada a uma base pequena ou moderada.


````python
for termo in ['backup', 'cabo', 'temperatura']:
    print(f"\nBusca por: {termo}")
    resultados = buscar_conhecimento(base, termo)
    pprint(resultados, sort_dicts=False)
````

### Analise da disseminacao

A busca implementada demonstra que o conhecimento explicito pode ser reutilizado por outros membros da equipe. Isso reduz tempo de resposta, melhora a padronizacao do atendimento e facilita a transferencia de conhecimento.


## 6. Parte 3 - Aplicacao por motor de regras

Na ultima fase, o sistema deixa de apenas armazenar e recuperar informacoes e passa a **aplicar conhecimento** em uma decisao automatica. A regra considera a temperatura do servidor para emitir um diagnostico operacional.


````python
for temperatura in [34.2, 35.7, 37.4]:
    print(f"{temperatura:.1f}C -> {diagnostico_servidor(temperatura)}")
````

### Analise da aplicacao

Esse motor de regras representa uma forma simples de inferencia. Mesmo sem inteligencia artificial avancada, o sistema ja transforma conhecimento codificado em acao recomendada ao usuario.
