## Objetivo

Simular o ciclo completo de um Data Warehouse (DW) por meio de um processo ETL (Extract, Transform, Load), integrando sistemas isolados e gerando uma "Verdade Única" organizacional.

# Contexto do Problema

A organização possui dois sistemas isolados:

- Sistema de Vendas → contém apenas SKU e quantidade vendida.
- Sistema de Produtos → contém SKU, nome e preço.

Nenhum dos sistemas, isoladamente, permite calcular o faturamento.

Este cenário caracteriza **Ilhas de Dados**.

```python
import pandas as pd

print("ETAPA 1 - EXTRAÇÃO")

vendas = pd.read_csv("vendas.csv")
produtos = pd.read_csv("produtos.csv")

print("\nTabela Vendas:")
display(vendas.head())

print("\nTabela Produtos:")
display(produtos.head())
```

## Interoperabilidade Técnica

Nesta etapa ocorre interoperabilidade técnica, pois arquivos CSV heterogêneos são lidos e interpretados por uma ferramenta comum (Python/Pandas).

```python
print("ETAPA 2 - TRANSFORMAÇÃO")

dw = pd.merge(vendas, produtos, on="sku")

print("\nTabela após Merge:")
display(dw.head())
```

## Interoperabilidade Semântica

O SKU atua como elo semântico entre os sistemas.

Sem um SKU padronizado, não seria possível associar corretamente quantidade e preço.

```python
dw["faturamento"] = dw["qtd"] * dw["preco"]

print("\nTabela com Faturamento:")
display(dw.head())
```

```python
dw["nome"] = dw["nome"].str.upper()

print("\nTabela após Padronização:")
display(dw.head())
```

```python
dw.to_csv("data_warehouse_final.csv", index=False)

print("Data Warehouse salvo com sucesso!")
```

# Conclusão

O processo ETL permitiu:

- Integrar silos organizacionais
- Aplicar regras de negócio centralizadas
- Garantir governança de dados
- Criar uma fonte única de verdade

O Data Warehouse consolidado permite análises estratégicas e decisões baseadas em dados.
