## Carregamento dos Dados

Nesta etapa, carregamos a base de conhecimento armazenada em JSON.
Caso o arquivo não exista, o sistema cria uma base inicial.

```python
from dados import carregar_dados, salvar_dados
from logica import buscar_conhecimento
from apresentacao import exibir_dashboard

base = carregar_dados()

display("Quantidade de registros:", len(base))
base
```

## Interface do Sistema (Menu Principal)

Abaixo está a implementação da interface baseada em menu, responsável pela interação com o usuário.

```python
def menu():
    base = carregar_dados()

    while True:
        display("\nSGC - MENU")
        display("1. Dashboard")
        display("2. Buscar")
        display("3. Adicionar")
        display("4. Sair")

        opcao = input("Escolha: ")

        if opcao == '1':
            exibir_dashboard(base)

        elif opcao == '2':
            termo = input("Busca: ")
            resultados = buscar_conhecimento(base, termo)

            display(f"\nResultados: {len(resultados)}")
            for r in resultados:
                display(f"{r['titulo']} -> {r['conteudo']}")
                r['acessos'] += 1

            salvar_dados(base)

        elif opcao == '4':
            salvar_dados(base)
            display("Encerrado.")
            break
```

## Busca de Conhecimento (Camada de Lógica)

A função implementa:
- Busca por título e conteúdo
- Case-insensitive
- Uso de List Comprehension

```python
display("\n>> Usuário escolheu: 2 (Buscar)")
display(">> Termo digitado: vpn\n")

resultados = buscar_conhecimento(base, "vpn")

display(f"Resultados: {len(resultados)}\n")

for r in resultados:
    display(f"{r['titulo']} -> {r['conteudo']}")
    r['acessos'] += 1
```

## Dashboard de Popularidade

Representação visual dos acessos usando barras.
Isso permite identificar rapidamente os conhecimentos mais utilizados.

```python
print("\n>> Usuário escolheu: 1 (Dashboard)\n")
exibir_dashboard(base)
```

## Inserção de Novo Conhecimento

Simulando a adição de um novo item na base.

```python
display("\n>> Usuário escolheu: 3 (Adicionar)\n")

novo = {
    "id": max([item['id'] for item in base]) + 1,
    "titulo": "Falha no Wi-Fi",
    "conteudo": "Reiniciar modem e verificar cabos.",
    "acessos": 0
}

base.append(novo)
salvar_dados(base)

display("Conhecimento adicionado com sucesso!")
```
